```python
import pandas as pd
import numpy as np
```

```python
titanic=pd.read_csv("train_data.csv")
df=titanic.copy()
df
```

```python
del df["Name"]
del df["Embarked"]
del df["Ticket"]
del df["Cabin"]
df
```

```python
def f(s):
    if s=="male":
        return 1
    else:
        return 0
df["Sex"]=df.Sex.apply(f)
df
```

```python
mean_s=df[df.Survived==1].Age.mean()
mean_u=df[df.Survived==0].Age.mean()
def fillnaage(df):
    for i in range(len(df.Age)):
        if np.isnan(df.Age[i]):
            if(df.Survived[i]==1):
                df.Age[i]=mean_s
            else:
                df.Age[i]=mean_u
    return df
df=fillnaage(df)
df.head(15)
```

```python
x_train=df.iloc[:,0:6]
y_train=df.iloc[:,6]
y_train,x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
alg=LogisticRegression()
```

```python
alg.fit(x_train,y_train)
```

```python
alg.score(x_train,y_train)
```
